```python
import pandas as pd
import matplotlib.pyplot as plt
#Loading the dataset
path = 'form.csv'
form_df = pd.read_csv(path)
form_df.dropna(axis=0, inplace=True)  # 删除包含NaN值的行
form_df.to_csv('form.csv', index=False)
print(form_df)

grouped_form_df = form_df.groupby('pet')
# 对pet类别进行分组
# 将分组后的数据写入新的csv文件中
for name, group in grouped_form_df:
    group.to_csv(name + '.csv', index=False)
```

```python
pa ='Dog.csv'
pb ='Cats.csv'
form_Dog = pd.read_csv(pa)
form_Cats = pd.read_csv(pb)
form_Dog['state'].value_counts().plot(kind='bar');
plt.title("Dog state")
plt.show()
form_Cats['state'].value_counts().plot(kind='bar');
plt.title("Cats state")
plt.show()
# 输出猫和狗的地区信息统计
```

```python
pa ='Dog.csv'
pb ='Cats.csv'
form_Dog = pd.read_csv(pa)
form_Cats = pd.read_csv(pb)
form_Dog['birth_month'].value_counts().plot(kind='bar');
plt.title("Dog birth_month")
plt.show()
form_Cats['birth_month'].value_counts().plot(kind='bar');
plt.title("Cats birth_month")
plt.show()
# 输出猫和狗的出生月份统计
```
